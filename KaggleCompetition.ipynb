```python
import numpy as np
import pandas as pd
import os

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, f1_score
```

```python
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train = pd.read_csv('/kaggle/input/competitions/ut-san-antonio-nlp/train.csv')
test = pd.read_csv('/kaggle/input/competitions/ut-san-antonio-nlp/test_solution.csv')
```

```python
train.columns = train.columns.str.strip()
test.columns = test.columns.str.strip()
```

```python
print(train.columns)
```

```python
for col in train.columns:
    print(repr(col))
```

```python
train['text'] = (
    "Question: " + train['interview_question'] +
    " Answer: " + train['interview_answer']
)

test['text'] = (
    "Question: " + test['interview_question'] +
    " Answer: " + test['interview_answer']
)
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    train['text'],
    train['clarity_label'],
    test_size=0.2,
    stratify=train['clarity_label'],
    random_state=42
)
```

```python
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(
        ngram_range=(1,2),
        max_features=30000,
        stop_words='english'
    )),
    ('clf', LogisticRegression(
        max_iter=2000,
        class_weight='balanced'
    ))
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
preds = pipeline.predict(X_val)


macro_f1 = f1_score(y_val, preds, average='macro')

print("Macro F1:", macro_f1)
print(classification_report(y_val, preds))
```
